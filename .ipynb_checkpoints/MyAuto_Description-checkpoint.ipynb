### 1.4 - Getting code ready

1) We will need to import numpy and opencv packages

```python
import numpy as np
import cv2
```

2) Lets create a `yolo_net`, `yolo_classes` and `yolo_outputlayers` variables which are used for openCV and YOLO connections

```python
yolo_net = cv2.dnn.readNet("../myauto_project_data/model_data/yolov3-tiny.weights", 
                           "../myauto_project_data/model_data/yolov3-tiny.cfg")
yolo_classes = []
with open("../myauto_project_data/model_data/coco.names", "r") as f:
    yolo_classes = [line.strip() for line in f.readlines()]
    
yolo_outputlayers = [yolo_net.getLayerNames()[i[0] - 1] 
                     for i in yolo_net.getUnconnectedOutLayers()]
```

3) Lets create a method `detect_car_on_image` that detects car(s) on pic and returns the highest percent of them

```python
def detect_car_on_image(img_path : str):
    img = cv2.imread(img_path)
    if(img is None):
        return None
    img = cv2.resize(img, (416,416))
    blob = cv2.dnn.blobFromImage(img, 0.004, (416, 416), (0,0,0), True, crop=False)
    yolo_net.setInput(blob)
    outs = yolo_net.forward(yolo_outputlayers)
    maxConf = 0
    for out in outs:
        for detection in out:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            if class_id == yolo_classes.index("car"):
                if maxConf < confidence:
                    maxConf = confidence
    return maxConf
```

### 1.5 - Implementing code that iters every pic in dataset

1) Lets create variables, where we will store some useful information. You **should** change path if you did not follow my instructions in **[1.1 - Getting Dataset Ready]**

In addition to this I created new directory "model_labels" ("../myauto_project_data/model_labels), where I will store the information about car pictures. We will use this variable lately in this part

```python
precode_dataset_path = "../myauto_project_data/images"
precode_characteristics_path =  "../myauto_project_data/characteristics.csv"
precode_car_picture_labels_path = "../myauto_project_data/model_labels/labels_all.csv"
```

2) Lets create a method that iters pics for car with id = `car_id` and returnes list of image pathes where YOLO detected a car

```python
import os
def get_car_pictures_from_id(car_id: int):
        car_dir_path = os.path.join(precode_dataset_path, str(car_id))
        if os.path.exists(car_dir_path):
            path, dirs, files = next(os.walk(car_dir_path))
            returnList = []
            for i, pic in enumerate(files):
                img_path = os.path.join(car_dir_path, pic)
                img_maxConf = detect_car_on_image(img_path)
                returnList.append((img_path, img_maxConf))
            return returnList
        else:
            return None
```

3) Lets read csv file using [pandas](https://pandas.pydata.org/) library.

```python
import pandas as pd
precode_characteristics = pd.read_csv(precode_characteristics_path)
```

4) Lets create method that iters every car in `precode_characteristics`, iters its images and assignes them the highest percent of the car, found in image. It returnes tuple of two list: `precode_car_id_and_yolo_output` which contains info about main data and `precode_columns` which contains column names in dataframe

```python
def create_data_for_labels_csv():
    precode_car_id_list = precode_characteristics.sample(frac=1)["ID"]
    precode_car_id_and_yolo_output = []
    precode_max_columns = 0
    for i, car_id in enumerate(precode_car_id_list):
        if(i%64 == 0):
            print("passed", i, "element")
        car_pictures = get_car_pictures_from_id(car_id)
        list_for_this_car = [car_id]
        if car_pictures is not None:
            if len(car_pictures) > precode_max_columns:
                precode_max_columns = len(car_pictures)
            for car_picture in car_pictures:
                percentage = car_picture[1]
                if (car_picture[1] is None) or (car_picture[1] is np.nan):
                    percentage = 0
                list_for_this_car.append(car_picture[0])
                list_for_this_car.append(int(percentage*100))
        precode_car_id_and_yolo_output.append(list_for_this_car)
    
    precode_columns = []
    for i in range(precode_max_columns):
        precode_columns.append("PIC" + str(i))
        precode_columns.append("CAR" + str(i))
        
    return precode_car_id_and_yolo_output, precode_columns
```

You can use your own machine to calculate and label every car picture using YOLO

### 2.1 - Getting dataset ready using csv file

#### 2.1.1 - reading information from csv file
In pervious part **[1 - Using YOLOv3-tiny to detect cars on data]** we labeled every picture we had in our Kaggle dataset and placed this information in csv file located on `precode_car_picture_labels_path`.

Firstly, we should load this information to pandas dataframe `car_picture_labels`.

Secondly, we should define variable `car_picture_label_columns`.

```python
car_picture_labels = pd.read_csv(precode_car_picture_labels_path)
car_picture_label_columns = car_picture_labels.columns[2:]
```

#### 2.1.2 - creating method to get the best pictures
In this part we will create a method which takes `least_percent` variable. It should iter every car picture which is labeled in `car_picture_labels`. For every picture It determines if there is higher chance than `least_percent` that this picture contains car. If so it adds this picture path and its color to the `color_dataframe_list`.

```python
def color_get_full_dataframe(least_percent: int):
    color_dataframe_list = []

    for index, row in car_picture_labels.iterrows():
        best_pic_pathes = []
        for column in car_picture_label_columns:
            if column.startswith("CAR"):
                current_percent = car_picture_labels.loc[index, column]
                if (current_percent is not np.nan) and (current_percent > least_percent):
                    best_pic_pathes.append(car_picture_labels.loc[index, "PIC" + column.split("CAR")[1]])

        id = row[2]
        color = precode_characteristics.loc[id, 'color']

        if color is not np.nan:
            for best_pic_path in best_pic_pathes:
                color_dataframe_list.append([best_pic_path, color]) 
    return color_dataframe_list
```

Now we need to get color classes into this variable `color_classes`

### 2.2 - Building model for color classification

In this part we will implement some cnn architecture for our model to detect cars and classify its colors. After some experiments I finally used architecture, which could detect colors on accuracy of 70%

#### 2.2.1 - building keras sequental architecture

For color classification I used Sequental architecture. Reason for this was that this kind of task (color classification) doesn't need some complex architectures to give good results.

I created **2 Convolution2D** layers (each containing **BatchNormalization** and **MaxPooling2D**), then Flattened their output to **3 Dense** layers and ended with **1 Dense softmax** layer.

For this we needed some imports from [Keras](https://keras.io/)

```python
from keras.models import Sequential
from keras.layers import BatchNormalization, Dense, Convolution2D, MaxPooling2D, AveragePooling2D, Dropout, Flatten


def color_net(num_classes):
    
    input_shape = (224, 224, 3)
    
    model = Sequential()
    model.add(Convolution2D(filters=128,kernel_size=(11,11),strides=(4,4),
                              input_shape=input_shape,activation='relu'))
    model.add(BatchNormalization())
    model.add(MaxPooling2D(pool_size=(3,3),strides=(2,2)))
    
    model.add(Convolution2D(filters=64, kernel_size=(3,3), strides=(1,1), activation='relu'))
    model.add(BatchNormalization())
    model.add(MaxPooling2D(pool_size=(3,3),strides=(2,2))) 

    model.add(Flatten())
    
    model.add(Dense(units=256, activation='relu'))
    model.add(Dropout(0.2))
    
    model.add(Dense(units=256, activation='relu'))
    model.add(Dropout(0.2))
    
    model.add(Dense(units=256, activation='relu'))
    model.add(Dropout(0.2))
    
    model.add(Dense(units=num_classes, activation='softmax'))
    
    return model
```

```python
category_train_datagen = ImageDataGenerator(
        rescale=1./255,
        width_shift_range=0.1,
        zoom_range = [0.6,0.7],
        brightness_range = (1.3, 1.7),
        shear_range=0.2,
        rotation_range = 40,
        horizontal_flip=True)

category_train_set = category_train_datagen.flow_from_dataframe(
         dataframe=category_dataframe_train,
         classes = category_classes,
         x_col = "file_path",
         y_col = "category",
         target_size=(category_img_rows, category_img_cols),
         batch_size=category_batch_size,
         class_mode='categorical')

category_val_set = category_train_datagen.flow_from_dataframe(
         dataframe=category_dataframe_val,
         classes = category_classes,
         x_col = "file_path",
         y_col = "category",
         target_size=(category_img_rows, category_img_cols),
         batch_size=category_batch_size,
         class_mode='categorical')
```

```python
category_first_batch = next(category_train_set)

category_key_list = list(category_train_set.class_indices.keys()) 
category_val_list = list(category_train_set.class_indices.values())

fig=plt.figure(figsize=(8, 8))
for i,x in enumerate(category_first_batch[0]):
    temp = fig.add_subplot(4, 4, i+1)
    plt.imshow(x)
    temp.set_title(category_key_list[category_val_list.index((np.argmax(category_first_batch[1][i])))])
    if(i>14):
        break
```

### 4.2 - Building model for category classification

In this part I used different architecture. It is called [MobileNet](https://keras.io/api/applications/mobilenet/). already trained model is ready to download on my [Google Drive](https://drive.google.com/file/d/14q-EJ9ppBxqm6YfILAnlcDL6cWN0P6xE/view?usp=sharing)

```python
def category_net(num_classes):   
    input_shape = (224, 224, 3)   
    
    model = MobileNet(input_shape=input_shape,
                        weights = None,
                        alpha = 0.5,
                        depth_multiplier=1,
                        dropout=0.01,
                        pooling = "max",
                        classes=num_classes)
      
    return model
```

```python
category_saved_model_path = "../myauto_project_data/model_saved/category_model_final.h5"

I_WANT_TO_TRAIN = False

if(I_WANT_TO_TRAIN):
    category_model = category_net(brand_num_classes)
else:
    category_model = load_model(category_saved_model_path)

category_model.summary()
```

### 4.3 - Training and Testing model

In this part we train our model (if `I_WANT_TO_TRAIN` is `True`) and test it on our test data

```python
if(I_WANT_TO_TRAIN):
    category_model.compile(optimizer="adam", loss='categorical_crossentropy', metrics=['accuracy'])
    category_model.fit_generator(category_train_set, epochs=30, validation_data=category_val_set)
    category_model.save('category_model_temp.h5')
```

```python
category_test_datagen = ImageDataGenerator(
        zoom_range = [0.6,0.7],
        brightness_range = (1.3, 1.7),
        rescale=1./255)

category_test_set = category_test_datagen.flow_from_dataframe(
         dataframe= category_dataframe_test,
         classes = category_classes,
         x_col = "file_path",
         y_col = "category",
         target_size=(category_img_rows, category_img_cols),
         batch_size=category_batch_size,
         class_mode='categorical')
```

```python
category_test_results = category_model.evaluate(category_test_set)
print("test loss, test acc:", category_test_results)
```
