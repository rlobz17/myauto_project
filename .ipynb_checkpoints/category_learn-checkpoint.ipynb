```python
import numpy as np
import pandas as pd
import Dataset_MyAuto
import matplotlib.pyplot as plt
import LabelCars
```

```python
images_dir = "../myauto_project_data/images"
info_dir = "../myauto_project_data/characteristics.csv"
dataset = Dataset_MyAuto.Dataset_MyAuto(images_dir, info_dir)
```

```python
carPicLabels = LabelCars.loadLabelsCSV()
carPicLabelsColumns = carPicLabels.columns[2:]
```

```python
dataframe_list = []

for index, row in carPicLabels.iterrows():
    least_percent = 85
    best_pic_pathes = []
    for column in carPicLabelsColumns:
        if column.startswith("CAR"):
            current_percent = carPicLabels.loc[index, column]
            if (current_percent is not np.nan) and (current_percent > least_percent):
                best_pic_pathes.append(carPicLabels.loc[index, "PIC" + column.split("CAR")[1]])

    id = row[2]
    category = dataset.getCategoryByID(id)

    if category is not np.nan:
        for best_pic_path in best_pic_pathes:
            dataframe_list.append([best_pic_path, category])
```

```python
category_classes = list(np.unique(np.array(dataframe_list)[:,1]))
print("we got", len(dataframe_list), "with",len(category_classes),"category_num")
```

```python
from sklearn.model_selection import train_test_split
train_and_val_dataframe_list, test_dataframe_list = train_test_split(dataframe_list, train_size = 0.8, random_state=2)
train_dataframe_list, val_dataframe_list = train_test_split(train_and_val_dataframe_list, train_size = 0.75, random_state=2)
print(len(dataframe_list), "splited to:", len(train_dataframe_list), len(val_dataframe_list), "and", len(test_dataframe_list))
```

```python
dataframe_train = pd.DataFrame(train_dataframe_list, columns=["file_path", "category"])
dataframe_test = pd.DataFrame(val_dataframe_list, columns=["file_path", "category"])
dataframe_final_test = pd.DataFrame(test_dataframe_list, columns=["file_path", "category"])
```

```python
img_rows , img_cols = 224,224
num_classes = len(category_classes)
batch_size = 16
nb_epoch = 15
```

```python
from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
        rescale=1./255,
        width_shift_range=0.1,
        zoom_range = [0.6,0.7],
        brightness_range = (1.3, 1.7),
        shear_range=0.2,
        rotation_range = 40,
        horizontal_flip=True)
# train_datagen = ImageDataGenerator(rotation_range=25, width_shift_range=0.1,
#     height_shift_range=0.1, shear_range=0.2, zoom_range=0.2,
#     horizontal_flip=True, fill_mode="nearest")
```

```python
train_set = train_datagen.flow_from_dataframe(
         dataframe=dataframe_train,
         classes = category_classes,
         x_col = "file_path",
         y_col = "category",
         target_size=(img_rows, img_cols),
         batch_size=batch_size,
         class_mode='categorical')

val_set = train_datagen.flow_from_dataframe(
         dataframe=dataframe_test,
         classes = category_classes,
         x_col = "file_path",
         y_col = "category",
         target_size=(img_rows, img_cols),
         batch_size=batch_size,
         class_mode='categorical')
```

```python
first_batch = next(train_set)

key_list = list(val_set.class_indices.keys()) 
val_list = list(val_set.class_indices.values())

fig=plt.figure(figsize=(8, 8))
for i,x in enumerate(first_batch[0]):
    temp = fig.add_subplot(4, 4, i+1)
    plt.imshow(x)
    temp.set_title(key_list[val_list.index((np.argmax(first_batch[1][i])))])
    if(i>14):
        break
```
